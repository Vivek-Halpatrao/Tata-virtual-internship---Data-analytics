### Filling Null values in Description using groupby().transform()
   As there is unique stockcode for every product so, replacing null values in description by using common stockcode

```python
df['final']=df.groupby('StockCode')['Description'].transform(lambda x: x.value_counts().keys().tolist()[0] if len(x.value_counts().keys().tolist())>0 else 'not defined' )
```

```python
df.drop(['Description'],inplace=True,axis=1)
```

### Found that Unit price for some items is zero and Quantity is negative which is technically not possible hence applying same groupby().transform method on Unit Price Columnn 

```python
df["new_unitprice"] = df.groupby("StockCode")['UnitPrice'].transform(lambda x: x.replace(0,x.mean()))
```
